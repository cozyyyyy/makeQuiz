```python
import pandas as pd
import json

jsonl_file = 'aio_04_dev_v1.0.jsonl'
csv_file = 'data.csv'

data_list = []

with open(jsonl_file, 'r') as file:
    for line in file:
        data = json.loads(line)
        data_list.append(data)


df = pd.DataFrame(data_list)

df.to_csv(csv_file, index=False)
```

```python
import os
import pandas as pd

df = pd.DataFrame(columns=['text', 'summary'])
with open('dict_data2.csv') as f:
    for line in f.readlines():
        strs = line.split(',')
        df = df.append({'text':strs[2] , 'summary':strs[0] + strs[1]}, ignore_index=True)

print(df)

df = df.sample(frac=1)

num = len(df)
df[:int(num*0.8)].to_csv('summary_train.csv', sep=',', index=False)
df[int(num*0.8):].to_csv('summary_val.csv', sep=',', index=False)
```

```python
import pandas as pd

# 空のデータフレームを作成
df = pd.DataFrame(columns=['text', 'summary'])

# ファイルを開いてデータを読み込む
with open('dict_data2.csv', 'r') as f:
    for line in f.readlines():
        strs = line.strip().split('\t')
        if len(strs) >= 2:
            new_row = {'text': strs[1], 'summary': strs[0]}
            df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

# データフレームの内容を表示
print(df)
```
